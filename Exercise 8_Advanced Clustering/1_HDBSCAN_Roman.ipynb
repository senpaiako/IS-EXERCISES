## I. HDBSCAN

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import hdbscan

%matplotlib inline
sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds = {'alpha': 0.5, 's': 80, 'linewidths': 0}
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df = df.iloc[:, [3, 4]].values
```

```python
# Function to plot clusters
def plot_clusters(clusterer, data, title):
    clusterer.fit(data)
    palette = sns.color_palette('deep', np.unique(clusterer.labels_).max() + 1)
    colors = [palette[x] if x >= 0 else (0.5, 0.5, 0.5) for x in clusterer.labels_]
    plt.scatter(data[:, 0], data[:, 1], c=colors, **plot_kwds)
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.title(title)

# Plot clusters for each set of parameters
plt.figure(figsize=(18, 6))

plt.subplot(131)
plot_clusters(hdbscan.HDBSCAN(min_cluster_size=3, gen_min_span_tree=True), df, 'Min Cluster Size = 3')

plt.subplot(132)
plot_clusters(hdbscan.HDBSCAN(min_cluster_size=5, gen_min_span_tree=True), df, 'Min Cluster Size = 5')

plt.subplot(133)
plot_clusters(hdbscan.HDBSCAN(min_cluster_size=9, gen_min_span_tree=True), df, 'Min Cluster Size = 9')

plt.show()
```
